By inspection, the 'Date' datatype is an object, we need to change it to datetime

```python
df['date'] = pd.to_datetime(df['Date'])
df['date'].dtype
type(df['date'])
```

```python
df['day'] = (df['date']).dt.day
df['month'] = (df['date']).dt.month
df['year'] = (df['date']).dt.year
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = (df['Time']).dt.hour    #type(sales['Time'])
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='Quantity',y='cogs',data=df,color='green')
plt.xlabel('Quantity')
plt.ylabel('Cost of Goods Sale')
plt.title('Quantity v Cost of Goods Sale',fontsize=15)
plt.show()
```
